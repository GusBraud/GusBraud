**Intégration d'un bloc de texte fixe et modification des lignes environnantes**

**Utilisation de la balise *persname* pour *lyricist***

```python
import os
import glob

input_folder = '/content/drive/MyDrive/Métadonnées_Gesualdo/Juillet_Août_2025/fichiers_MEI_versionfinale_200825_Out'
output_folder = '/content/drive/MyDrive/Métadonnées_Gesualdo/Juillet_Août_2025/fichiers_MEI_versionfinale_010925_Out'

# Create the output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

mei_paths = glob.glob(os.path.join(input_folder, '*'))

for mei_path in mei_paths:
    if os.path.isfile(mei_path):
        print(f"Processing file: {mei_path}")
        with open(mei_path, 'r', encoding='utf-8') as f:
            lines = f.readlines()

        modified_lines = []
        for line in lines:
            if '<lyricist xml:id=' in line and '</lyricist>' in line:
                # Extract the content within the lyricist tag
                start_index = line.find('>') + 1
                end_index = line.rfind('<')
                lyricist_content = line[start_index:end_index]

                # Construct the new line with the desired semantic
                modified_line = line.replace(line.strip(), f'<lyricist>\n            <persName {line[line.find("xml:id="):line.find(">")]}>{lyricist_content} </persName>\n          </lyricist>')
                modified_lines.append(modified_line)
            else:
                modified_lines.append(line)

        # Create the output file path
        output_path = os.path.join(output_folder, os.path.basename(mei_path))
        print(f"Saving modified file to: {output_path}")

        # Write the modified content to the output file
        with open(output_path, 'w', encoding='utf-8') as f:
            f.writelines(modified_lines)
    else:
        print(f"Skipping directory: {mei_path}")

print("Lyricist metadata updated and files saved to:", output_folder)
```

**Utilisation de la balise *persname* pour *dedicatee***

```python
import os
import glob

input_folder = '/content/drive/MyDrive/Métadonnées_Gesualdo/Juillet_Août_2025/fichiers_MEI_versionfinale_010925_Out'
output_folder = '/content/drive/MyDrive/Métadonnées_Gesualdo/Juillet_Août_2025/fichiers_MEI_versionfinale_010925_Out_Dedicattee'

# Create the output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

mei_paths = glob.glob(os.path.join(input_folder, '*'))

for mei_path in mei_paths:
    if os.path.isfile(mei_path):
        print(f"Processing file: {mei_path}")
        with open(mei_path, 'r', encoding='utf-8') as f:
            lines = f.readlines()

        modified_lines = []
        for line in lines:
            if '<dedicatee xml:id=' in line and '</dedicatee>' in line:
                # Extract the content and attributes within the dedicatee tag
                start_content_index = line.find('>') + 1
                end_content_index = line.rfind('<')
                dedicatee_content = line[start_content_index:end_content_index].strip()

                start_attr_index = line.find('xml:id=')
                end_attr_index = line.find('>', start_attr_index)
                dedicatee_attributes = line[start_attr_index:end_attr_index]


                # Construct the new line with the desired semantic
                modified_line = line.replace(line.strip(), f'<dedicatee>\n            <persName {dedicatee_attributes}>{dedicatee_content} </persName>\n          </dedicatee>')
                modified_lines.append(modified_line)
            else:
                modified_lines.append(line)

        # Create the output file path
        output_path = os.path.join(output_folder, os.path.basename(mei_path))
        print(f"Saving modified file to: {output_path}")

        # Write the modified content to the output file
        with open(output_path, 'w', encoding='utf-8') as f:
            f.writelines(modified_lines)
    else:
        print(f"Skipping directory: {mei_path}")

print("Dedicatee metadata updated and files saved to:", output_folder)
```

**Correction de l'indentation et suppression d'éventuelles erreurs à la fin du document**

```python
import os
import glob
import xml.dom.minidom # Using a standard XML library for indentation

input_folder = '/content/drive/MyDrive/Métadonnées_Gesualdo/Juillet_Août_2025/fichiers_MEI_versionfinale_010925_Out_Dedicattee'
output_folder = '/content/drive/MyDrive/Métadonnées_Gesualdo/Juillet_Août_2025/fichiers_MEI_versionfinale_010925_Out_Indented'
example_mei_path = '/content/drive/MyDrive/Métadonnées_Gesualdo/Juillet_Août_2025/Ha_ha_que_de_bien_a4_UhgyHoA.mei' # Path to the example file for indentation

# Create the output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

mei_paths = glob.glob(os.path.join(input_folder, '*'))

# Determine indentation from the example file (optional, but can help match style)
# This is a simple approach and might need refinement based on the exact desired style
example_indent = "    " # Default to 4 spaces if example parsing fails or isn't desired
try:
    with open(example_mei_path, 'r', encoding='utf-8') as f:
        example_lines = f.readlines()
        for line in example_lines:
            if line.strip() and not line.strip().startswith('<'): # Find a line with content but not an opening tag
                leading_whitespace = line[:len(line) - len(line.lstrip())]
                if leading_whitespace:
                    example_indent = leading_whitespace
                    break
except Exception as e:
    print(f"Could not determine indentation from example file: {e}. Using default indentation.")

for mei_path in mei_paths:
    if os.path.isfile(mei_path):
        print(f"Processing file for indentation and trimming: {mei_path}")
        try:
            with open(mei_path, 'r', encoding='utf-8') as f:
                lines = f.readlines()

            # Remove lines after </mei>
            mei_end_index = -1
            for i, line in enumerate(lines):
                if line.strip() == "</mei>":
                    mei_end_index = i
                    break

            if mei_end_index != -1:
                lines = lines[:mei_end_index + 1] # Keep the </mei> line and everything before it


            # Re-indent the XML content
            content = "".join(lines)
            # Use minidom to parse and re-indent
            dom = xml.dom.minidom.parseString(content)
            # prettyprint with custom indentation (minidom uses tabs by default, this is a workaround)
            # A more robust solution for specific indentation might require a different library or manual processing
            indented_content = '\n'.join([line for line in dom.toprettyxml(indent=example_indent).split('\n') if line.strip()])


            # Create the output file path
            output_path = os.path.join(output_folder, os.path.basename(mei_path))
            print(f"Saving modified file after indentation and trimming to: {output_path}")

            # Write the modified content to the output file
            with open(output_path, 'w', encoding='utf-8') as f:
                f.write(indented_content)

        except Exception as e:
            print(f"Error processing file {mei_path}: {e}")

    else:
        print(f"Skipping directory: {mei_path}")

print("Files indented and trimmed, saved to:", output_folder)
```
